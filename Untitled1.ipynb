```python
#mutual info based feature selec

import pandas as pd
mydata = 'GSE64810'
data_file = './data/'+ mydata +'_Train.csv'
df=pd.read_csv(data_file).T

new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header

df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels=['disease'], axis=1),
    df['disease'].astype('str'),
    test_size=0.3,
    random_state=0)
from sklearn.feature_selection import mutual_info_classif
mutual_info = mutual_info_classif(X_train, y_train)
```

```python
mutual_info1 = pd.Series(mutual_info)
mutual_info1.index = X_train.columns
TopMarker=mutual_info1.sort_values(ascending=False)
```

```python
TopMarker[1:100].plot(kind='bar',alpha=0.75,rot=90,figsize=(20,6))
```
